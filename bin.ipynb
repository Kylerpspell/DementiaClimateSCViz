```python
import csv
with open('data/zip2County.csv', 'r') as f:
	reader = csv.reader(f)
	zip2County = {rows[0]:rows[1] for rows in reader}
```

```python
with open('data/UseableData.csv', 'r') as f:
	reader = csv.reader(f)
	header = next(reader)
	data = list(reader)
```

```python
zipbin = {}
for row in data:
	try:
		zip = row[18]
		if zip2County[zip] not in zipbin:
			zipbin[zip2County[zip]] = 1
		else:
			zipbin[zip2County[zip]] += 1
	except:
		print('Error in' +  row[22])
		pass

with open('data/zipbin.csv', 'w') as f:
	writer = csv.writer(f)
	writer.writerow(['County', 'Count'])
	for key, value in zipbin.items():
		writer.writerow([key, value])
```

```python
with open('countypop.csv','r') as f:
	reader = csv.reader(f)
	header = next(reader)
	popdata = list(reader)
```

```python
cleanedPopData = []

for row in popdata:
	cleanedPopData.append([row[0].replace(' County, South Carolina','').lstrip('.'),int(row[1])])
```

```python
print(cleanedPopData)
```

```python
for row in cleanedPopData:
	try:
		zipbin[row[0]] = zipbin[row[0]]/row[1]
	except:
		print('Error in' +  row[0])
		pass
```
